latihan 1 (Rektor BFS)

```python
def bfs(grafik, start):
    visited = []  
    queue = []    
    queue.append(start)
    visited.append(start)
    
    while queue:
        current = queue.pop(0)
        for neighbor in grafik[current]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)
    return visited
grafik2 = {
    'Rektor': ['Warek 1', 'Warek 2'],
    'Warek 1': [],
    'Warek 2': ['Kaprodi 1', 'Kaprodi 2', 'Kaprodi 3'],
    'Kaprodi 1': ['Dosen A', 'Dosen B', 'Dosen C'],
    'Kaprodi 2': ['Dosen D', 'Dosen E'],
    'Kaprodi 3': ['Dosen F', 'Dosen G'],
    'Dosen A': [],
    'Dosen B': [],
    'Dosen C': [],
    'Dosen D': [],
    'Dosen E': [],
    'Dosen F': [],
    'Dosen G': []
}

bfs_method = bfs(grafik2, 'Rektor')
print("Visited nodes:", bfs_method)
```

latihan 2 (Rektor DFS)

```python
def dfs(graph, start):
    visited = []
    stack = []
    stack.append(start)
    while stack:
        current = stack.pop()
        if current not in visited:
            visited.append(current)
            for neighbor in reversed(graph[current]):
                if neighbor not in visited:
                    stack.append(neighbor)
    return visited
dfs_method = dfs(grafik2,'Rektor')
print("Visited nodes:", dfs_method)
```

Amin bfs & dfs

```python
Graph = {
    'Amin': ['Wasim', 'Nick', 'Mike'],
    'Wasim': ['Imran'],
    'Imran': ['Faras'],
    'Nick' : [],
    'Mike': [],
    'Faras': []
}
dfs_method = dfs(Graph,'Amin')
bfs_method = bfs(Graph,'Amin')
print("DFS Visited nodes:", dfs_method)
print("BFS Visited nodes:", bfs_method)
```
